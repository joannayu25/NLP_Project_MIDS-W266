```python
domain_relevance_df = pd.read_pickle('/content/drive/My Drive/W266/Project/pickle/domain_relevance_df.pkl')
domain_relevance_df 
```

```python
labels = ['20 Labeled Examples', '200 Labeled Examples', '2000 Labeled Examples', '5000 Labeled Examples']
imdb = domain_relevance_df[domain_relevance_df['Dataset']=='IMDb Movie Reviews']['Error Rate']
am_mov = domain_relevance_df[domain_relevance_df['Dataset']=='Amazon Movie and TV Reviews']['Error Rate']
am_ofc = domain_relevance_df[domain_relevance_df['Dataset']=='Amazon Office Product Reviews']['Error Rate']
airline = domain_relevance_df[domain_relevance_df['Dataset']=='Twitter Airline Sentiment']['Error Rate']
nat_dis = domain_relevance_df[domain_relevance_df['Dataset']=='Kaggle Natural Disaster Tweets']['Error Rate']

x = np.arange(len(labels))  # the label locations
width = 0.16  # the width of the bars

fig, ax = plt.subplots(figsize=(20, 10))
rects1 = ax.bar(x - width*2, imdb, width, color='navy', label='IMDb Movie Reviews')
rects2 = ax.bar(x - width, am_mov, width, color='mediumblue', label='Amazon Movie and TV Reviews')
rects3 = ax.bar(x, am_ofc, width, color = 'royalblue', label='Amazon Office Product Reviews')
rects4 = ax.bar(x + width, airline, width, color = 'deepskyblue', label='Twitter Airline Sentiment')
rects5 = ax.bar(x + width*2, nat_dis, width, color = 'powderblue', label='Kaggle Natural Disaster Tweets')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Error Rate', fontsize=20, labelpad=30)
ax.set_title('Performance Comparison for Different Domain', fontsize=22,fontweight='bold', pad=30)
ax.set_xticks(x)
ax.tick_params(axis='y', labelsize=18)
ax.set_xticklabels(labels, fontsize=18)
ax.legend(fontsize=20)

def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{:.3f}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')

autolabel(rects1)
autolabel(rects2)
autolabel(rects3)
autolabel(rects4)
autolabel(rects5)

plt.savefig('/content/drive/My Drive/W266/Project/images/domain_relevance_blue.png')
plt.show()
```
